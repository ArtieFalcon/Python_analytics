```python
#gender,"race/ethnicity","parental level of education","lunch","test preparation course","math score","reading score","writing score"
with open('student_performance.csv') as f:
    students_info={}
    gender={}
    race={}
    lunch={}
    math_score_gender=math_score_r=math_score_l=0
    i=0
    j=0
    k=0
    for line in f:
        info=line.strip().split(',')
        if 'gender' in info[0]:
            continue
        elif info[0][1:-1] == 'male':# and info[1][1:-1] == 'group B'and info[3][1:-1] = 'standard':
            i+=1
            math_score_gender+=int(info[-3][1:-1])
        if info[1][1:-1] == 'group B':
            j+=1
            math_score_r+=int(info[-3][1:-1])
        if info[3][1:-1] == 'standard':
            k+=1
            math_score_l+=int(info[-3][1:-1])
    gender['count']=i
    gender['math_score_mean']=round(math_score_gender/i,3)
    race['count']=j
    race['math_score_mean']=round(math_score_r/j,3)
    lunch['count']=k
    lunch['math_score_mean']=round(math_score_l/k,3)
    
    students_info['gender']=gender
    students_info['race']=race
    students_info['lunch']=lunch
    print(students_info)
   
```

```python
div_list = [12, 5, 0, 16, 7]
def division_18(div_list):
    result=[]
    for x  in div_list:
        try:
            result.append(round(18/x,3))
        except ZeroDivisionError:
            result.append(0)
    return result
division_18(div_list)    
```

#Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.

```python
s= "([)][()](())" #"([)]" #"([)]" #'(){}{([}])[]{{()[]}}' #[)(]
def isValid(s):
    d={}
    d['(']=1
    d[')']=-1
    d['{']=2
    d['}']=-2
    d['[']=5
    d[']']=-5

    slist=list(s)
    slist_values=[]
    for i in slist:
        for k,v in d.items():
            if i==k:
                slist_values.append(v)
    #print(slist_values)
    if sum(slist_values)!=0 or len(slist_values)%2==1:
        return False
    else:
        summ=0
        for j in slist_values:
                summ+=j
                if summ<0:
                    return False
        #listk= [i * (-1) for i in slist_values] #generate multiplying list        
    slist_values2=slist_values
    c=0
    while c==0:
        le=len(slist_values2)
        for k in range(0,le-1):
            if slist_values2[k]+slist_values2[k+1]==0:
                slist_values2[k]=0
                slist_values2[k+1]=0
        if 0 in slist_values2:
            slist_values2=list(filter(lambda x: x!=0, slist_values2))
        elif len(slist_values2)==0:  
            c=1
            return True
        elif len(slist_values2)!=0:  
            c=1
            return False
       
    return True
print(isValid(s))
```

```python
nums = [1,3,5,6]
target = 4
def searchInsert(nums, target):
    if target in nums:
        return nums.index(target)
    else:
        i=0
        while i <len(nums):
            if target > nums[i]:
                i+=1
            else:
                return i
    return i
searchInsert(nums, target)        
        
```
