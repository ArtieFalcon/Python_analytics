7. Дан массив arr, необходимо найти среднее значение внутри групп, группа определена массивом g.
Например, для
arr = [4, 2, 1, 5, 2, 5, 1, 3] и
g =   [0, 1, 1, 0, 1, 0, 1, 0] (4 относится к группе 0, 2 к группе 1 и т. д.)
Средние по группам: [4.25, 1.5]
Подсказка: np.bincount()
Используйте библиотеку numpy, не используйте циклы.

```python
arr = [4, 2, 1, 5, 2, 5, 1, 3] # вес
g =   [0, 1, 1, 0, 1, 0, 1, 0] #группы
qty_0=(g.count(g[0])) #кол-во элементов первой группы (группа 0) - группа с которой начинается список g
qty_1=list(filter(lambda x: x!= g[0], g)).count(list(filter(lambda x: x!= g[0], g))[0]) 
# - кол-во элементов второй группы (группа НЕ 0 - то есть 1) 
# - т.е. группа с которой не начинается список g. Возможно явное определение вроде qty_1=g.count(1) 
#- но такое определение неуниверсально, т.к. вместо 1 могло стоять и другое число, т.е. третья группа
print('элементов в группах:',qty_0, qty_1)
s=np.bincount(g,  weights=arr)
print(s) #-сумма
means=np.array([s[0]/qty_0, s[1]/qty_1]) #рассчет среднего
print('среднее: ',means)
```
